**Data Cleaning**

```python
main_df.isnull().sum() 
```

```python
#here we are going to drop unecessary columns and columns which have many missing values

main_df.drop('TAIL_NUMBER' , axis = 1 , inplace = True)
main_df.drop('DEPARTURE_TIME' , axis = 1 , inplace = True)
main_df.drop('DEPARTURE_DELAY' , axis = 1 , inplace = True)
main_df.drop('TAXI_OUT' , axis = 1 , inplace = True)
main_df.drop('WHEELS_OFF' , axis = 1 , inplace = True)
main_df.drop('ELAPSED_TIME' , axis = 1 , inplace = True)
main_df.drop('AIR_TIME' , axis = 1 , inplace = True)
main_df.drop('WHEELS_ON' , axis = 1 , inplace = True)
main_df.drop('TAXI_IN' , axis = 1 , inplace = True)
main_df.drop('CANCELLATION_REASON' , axis = 1 , inplace = True)
main_df.drop('AIR_SYSTEM_DELAY' , axis = 1 , inplace = True)
main_df.drop('SECURITY_DELAY' , axis = 1 , inplace = True)
main_df.drop('AIRLINE_DELAY' , axis = 1 , inplace = True)
main_df.drop('LATE_AIRCRAFT_DELAY' , axis = 1 , inplace = True)
main_df.drop('WEATHER_DELAY' , axis = 1 , inplace = True)

```

```python
main_df.isnull().sum()
```

```python
#Detecting and Removing Outliers
#box-plot before removing outliers
sns.boxplot(x=main_df['LATITUDE'])
```

```python
Q1 = main_df['LATITUDE'].quantile(0.25)
Q3 = main_df['LATITUDE'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
arr_1 = main_df['LATITUDE'] < (Q1 - 1.5 * IQR)
arr_2 = main_df['LATITUDE'] > (Q3 + 1.5 * IQR)
mask = (arr_1 | arr_2)
```

```python
print("No. of outliers: ", mask.sum())
```

```python
#Removing the outliers
main_df = main_df[~(mask)]
main_df.head()
```

```python
#box-plot after removing outliers
sns.boxplot(x=main_df['LATITUDE'])
```

```python
#box-plot before removing outliers
sns.boxplot(x=main_df['LONGITUDE'])
```

```python
Q1 = main_df['LONGITUDE'].quantile(0.25)
Q3 = main_df['LONGITUDE'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
arr_1 = main_df['LONGITUDE'] < (Q1 - 1.5 * IQR)
arr_2 = main_df['LONGITUDE'] > (Q3 + 1.5 * IQR)
mask = (arr_1 | arr_2)
```

```python
print("No. of outliers: ", mask.sum())
```

```python
#Removing the outliers
main_df = main_df[~(mask)]
main_df.head()
```

```python
#box-plot after removing outliers
sns.boxplot(x=main_df['LONGITUDE'])
```

```python
#fill the missing values for numerical terms - mean(Continous Variables) - after removing outliers
main_df['LONGITUDE'] = main_df['LONGITUDE'].fillna(main_df['LONGITUDE'].mean())
main_df['LATITUDE'] = main_df['LATITUDE'].fillna(main_df['LATITUDE'].mean())
```

**Feature Engineering & Visualizations**

```python
df_main_demo2=main_df
```

```python
df_main_demo2.head()
```

```python
#Newley Created Feature 'DATE' that will be calculated using the features 'YEAR','MONTH', 'DAY'
df_main_demo2['DATE'] = pd.to_datetime(df_main_demo2[['YEAR','MONTH', 'DAY']])
```

```python
#here this features helps me calculate the passenger footstep in each given airport 
air_trips = df_main_demo2.groupby(['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'], as_index=False)['FLIGHT_NUMBER'].count()
air_trips.sort_values(by='FLIGHT_NUMBER',inplace=True, ignore_index=True)
air_trips['Trips'] = air_trips.apply(lambda x: str(x['ORIGIN_AIRPORT'])+'-'+str(x['DESTINATION_AIRPORT']),axis=1)
sns.catplot(y='Trips', x='FLIGHT_NUMBER', kind='bar', 
            data=air_trips.iloc[-10:,:], 
            aspect=2)
plt.xlabel('Number of Flights')
plt.ylabel('Trips')
plt.show()
```

```python
df_main_demo2.head(1)
```

```python
#here i created a new feature (hour) that will help me calculate the number of flights per hour
#as demonstrated in the visualization below
#this new feature was created to answer the question of ( What are the hours of the days that have the most flights)

df_main_demo2['HOUR']=(df_main_demo2.SCHEDULED_DEPARTURE/100).astype(int)
```

```python
#The bar plot, clearly shows how, regardless of the month, 
#the number of flights drops down at night 
#and it reaches is maximim values between 6 and 7 AM, 
#with other two small peaks between 15 and 16, and 17 and 18.

pddf=pd.DataFrame({'count' : df_main_demo2.groupby(((df_main_demo2.SCHEDULED_DEPARTURE/100).astype(int))).size()}).reset_index()

# plot the number of flights per hour
f, ax = plt.subplots(figsize=(18, 6))
sns.barplot(x="SCHEDULED_DEPARTURE",
            y="count",
            data=pddf,
            palette=sns.color_palette("GnBu_d", 24),
            ax=ax)

# set plot's labels
ax.set(xlabel='Hour of the day', ylabel='Number of flights')
ax.set_title("Number of flights per hour during the year", fontweight="bold", size=15)

plt.plot()
```

```python
#Here we're trying to add a new feature "Percentage_per_hour" to know the percentage of flights delays occur through the hours of a day 
#and then determine the hours with the highest and lowest delay percemtages.
# A flight is considered to be delayed if the arrival delay(ARRIVAL_TIME-SCHEDULED_ARRIVAL) is greater than 15
#we're creating a new dataframe to count the number of delayed flights for each hour
delay_per_hour_DF = pd.DataFrame({'count' : df_main_demo2[df_main_demo2.ARRIVAL_DELAY > 15].groupby('HOUR').size()}).reset_index()

```

```python
# we're creating a new dataframe to count the number of flights for each hour
flights_per_hour_DF = pd.DataFrame({'count' : df_main_demo2.groupby('HOUR').size()}).reset_index()
```

```python
#Here we're joining the 2 dataframes created using Hour Column to be able to calculate our new feature "Percentage_per_hour"
#by dividing the count of the delayed flights per hour we got over the count of all flights per hour multiplied by 100 to get the percentage 
#and then adding the new feature with the hour column in a new "Percentage_hour_delay" dataframe to be able 
#to visualize and answer the research Question. 
df=flights_per_hour_DF.join(delay_per_hour_DF,on='HOUR',rsuffix='_d', how='inner')
df['Percentage_per_hour']=df['count_d']*100/df['count']
Percentage_hour_delay = df[['HOUR','Percentage_per_hour']]
```

```python
Percentage_hour_delay.head(10)
```

```python
#here the Graph shows that 5 AM is the time with lowest flight delay percentages in opposition to 6,7 and 8 PM 
#which are the peak of the flight delay percentages.
f, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x="HOUR", y="Percentage_per_hour", data=Percentage_hour_delay, palette=sns.color_palette("viridis", 24), ax=ax)
#labels
ax.set(xlabel='Hour of the day', ylabel='Percentage of delayed flights')
ax.set_title("Percentage of delayed flights for different hours of the day", fontweight="bold", size=15)
plt.plot()
```

```python
#Here we're trying to add a new feature "Percentage_Per_Month" to know the percentage of flights delays occur through the months of a year 
#and then determine the months with the highest and lowest delay percemtages.
#A flight is considered to be delayed if the arrival delay(ARRIVAL_TIME-SCHEDULED_ARRIVAL) is greater than 15.
#we're creating a new dataframe to count the number of delayed flights for each month.
Delay_per_Month_DF =  pd.DataFrame({'count' : df_main_demo2[df_main_demo2.ARRIVAL_DELAY > 15].groupby('MONTH').size()}).reset_index()
```

```python
#we're creating a new dataframe to count the number of flights for each month
flights_per_month_DF= pd.DataFrame({'count' : df_main_demo2.groupby('MONTH').size()}).reset_index()
```

```python
#here we're creating our new feature "Percentage_Per_Month" by dividing the count of the delayed flights per Month we got over the count of all flights per Month 
#multiplied by 100 to get the percentage and then adding the new feature to the dataframe to be able 
#to visualize and answer the research Question. 
Delay_per_Month_DF['Percentage_Per_Month']=Delay_per_Month_DF['count']*100/flights_per_month_DF['count']
```

```python
Delay_per_Month_DF.head(13)
```

```python
#here the Graph shows that june & february is the time with higest flight delay percentages in opposition to september which has the lowest
#flight delay percentages.
f, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x="MONTH",
            y="Percentage_Per_Month",
            data=Delay_per_Month_DF,
            palette=sns.color_palette("rocket", 12),
            ax=ax)

# selabels
ax.set(xlabel='Month', ylabel='Percentage of delayed flight')
#months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November','December']
#ax.set_xticklabels(months)
ax.set_title("Percentage of delayed flights over the months of the year", fontweight="bold", size=15)

plt.plot()
```
